```
valley=[]
print('test')
count=0
for i in range(119909):
    if(b[i]>b[i+1]):
        i=i+1
    elif(b[i]<b[i+1]):
        valley.append(b[i])
        i=i+1
        count=count+1
type(valley)
print(count)
```

```
def countlist(valley,peak):
    return[sub[item] for item in range(len(valley))
                     for sub in[valley,peak]]

com=(countlist(valley,peak))
e=len(com)
print(com)
```

```
cfc=0
fc=[]
lam=[]
lc=[]
ms=[]
for i in range(101713):
    m=com[i]*C
    n=com[i+1]*C
    o=m*n
    p=C1*com[i-1]
    if(o<0):
        cfc=cfc+1
        if(cfc==2):
            cfc=0
            if(i%4==0):
                fc.append(com[i])
            elif(i%4==1)and(p<=com[i]<=C) :
                lam.append(com[i])
            elif(i%4==2)and(com[i]<C2):
                lc.append(com[i])
            elif(i%4==3):
                ms.append(com[i])
print('FIRST TOUCH')
print(fc)
print('LAMBDA')
print(lam)
print('LAST TOUCH')
print(lc)
print('MID SWING')
print(ms)        
```
